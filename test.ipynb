```python
def f(x):
    return x * x + 1.0


plot_function("f(x)", f)
```

```python
def d_f(x):
    return 2 * x


def tangent_line(slope, x, y):
    def line(x_):
        return slope * (x_ - x) + y

    return line


plot_function("f(x) vs f'(2)", f, fn2=tangent_line(d_f(2), 2, f(2)))
```

```python
def sigmoid(a):
    return 1 / (1+math.exp(-a))

plot_function("sigmoid", sigmoid)
```

```python
from minitorch.scalar_functions import ScalarFunction

class Sin(ScalarFunction):
    @staticmethod
    def forward(ctx, x):
        ctx.save_for_backward(x)
        return math.sin(x)

    @staticmethod
    def backward(ctx, d):
        (x,) = ctx.saved_values
        f_prime = math.cos(x)
        return f_prime * d
    
def d_call(x):
    ctx = minitorch.Context()
    Sin.forward(ctx, x)
    return Sin.backward(ctx, 1)


plot_function("f(x) = sin(x)", lambda x: Sin.apply(x).data)
plot_function("1 * f'(x) = cos(x)", d_call)
```

topo sort

```python
root = {
    "data": 1,
    "parents": [{
        "data": 2,
        "parents": [{
                "data": 5,
            },{
                "data": 6,
                "parents": [{
                    "data": 7,
                },{
                    "data": 8,
                }]
            }]
        },{
            "data": 3,
        },{
            "data": 4,
        }]
}

def topo_sort(root):    
    parents = []
    
    for parent in root.get('parents', []):
        parents.extend(topo_sort(parent))
       
    parents.append(root)         
    return parents

res = []
res.extend(topo_sort(root))
for item in res:
    print(item['data'])
```
